## Jam e Hazfi

### Teams

```python
def get_score(x):
    if '(' not in x:
        return x
    flag = False
    end, start = len(x), 0
    for i in range(len(x)):
        if x[i] == '(' and i != 0: 
            return x[start:i]
        elif x[i] == '(' and i == 0:
            flag = True
        elif flag and x[i] == ')':
            return x[i+1:]
        
        
def get_penalty(x):
    if '(' not in x:
        return 0
    pattern = re.compile(r"\((\d+)\)")
    return pattern.findall(x)[0]
        
def get_data(x, html_tag, team_num=None):
    team = ''
    if team_num == 1:
        team = 'left'
    elif team_num == 2:
        team = 'right'
    if html_tag == 'penalty':
        html_tag = 'team-score '
        return get_penalty(BeautifulSoup(x, 'html.parser').find_all(class_=html_tag + team)[0].get_text())
    elif html_tag == 'team-score ':
        return get_score(BeautifulSoup(x, 'html.parser').find_all(class_=html_tag + team)[0].get_text())
    else:
        return BeautifulSoup(x, 'html.parser').find_all(class_=html_tag + team)[0].get_text()
    
```

```python
team_tag = 'team-name '
jame_hazfi['team1'] = jame_hazfi['html'].apply(lambda x: get_data(x, team_tag, 1))
jame_hazfi['team2'] = jame_hazfi['html'].apply(lambda x: get_data(x, team_tag, 2))
```

### Scores

```python
score_tag = 'team-score '
jame_hazfi['score1'] = jame_hazfi['html'].apply(lambda x: get_data(x, score_tag, 1))
jame_hazfi['score2'] = jame_hazfi['html'].apply(lambda x: get_data(x, score_tag, 2))
```

### Match Time

```python
time_tag = 'match-time pull-right'
jame_hazfi['time'] = jame_hazfi['html'].apply(lambda x: get_data(x, time_tag))
```

### Match Date

```python
date_tag = 'match-date pull-right'
jame_hazfi['date'] = jame_hazfi['html'].apply(lambda x: get_data(x, date_tag))
```

### Penalties

```python
penalty_tag = 'penalty'
jame_hazfi['penalty1'] = jame_hazfi['html'].apply(lambda x: get_data(x, penalty_tag, 1))
jame_hazfi['penalty2'] = jame_hazfi['html'].apply(lambda x: get_data(x, penalty_tag, 2))
```

```python
tidy_jame_hazfi = jame_hazfi.drop(columns=['Unnamed: 0', 'html'])
tidy_jame_hazfi.to_csv('Tidy Data/tidy_jame_hazfi.csv')
```

```python
def get_tables_data(leagues):
    new_df = pd.DataFrame()
    for competition in leagues:
        competitions = tables[tables['Competition'] == competition].reset_index(drop=True)
        for i in range(competitions.shape[0]):
            temp = pd.read_html(competitions['html'][i], header=1)[0] 
            if 'Unnamed: 10' in temp.columns:
                temp = temp.drop(columns=['Unnamed: 10', 'Unnamed: 0'])
            temp['competition'] = competition
            new_df = new_df.append(temp)
    return new_df.reset_index(drop=True)
```

```python
new_basketball = get_tables_data(basketball)
new_volleyball = get_tables_data(volleyball)
new_football_footsal = get_tables_data(football_footsal)
new_football_footsal
```

```python
new_basketball.drop(columns=['Unnamed: 1', 'Unnamed: 9'], inplace=True)
new_basketball
```

```python
new_volleyball.drop(columns=['Unnamed: 0'], inplace=True)
new_volleyball
```

```python
def get_team_data(x, data_type):
    data = BeautifulSoup(x, 'html.parser').find('tr', class_='highlight')
    if data is not None:
        if data_type == 'score':
            return int(data.contents[-4].text)
        elif data_type == 'matches':
            return int(data.contents[-2].text)
    else:
        return 0
```

```python
teams['score'] = teams['html'].apply(lambda x: get_team_data(x, 'score'))
teams['matches'] = teams['html'].apply(lambda x: get_team_data(x, 'matches'))
```

```python
tidy_teams = teams.drop(columns=['Unnamed: 0', 'html'])
tidy_teams.to_csv('Tidy Data/tidy_teams.csv')
tidy_teams
```
